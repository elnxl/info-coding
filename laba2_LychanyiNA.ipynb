```python
import numpy as np
from math import exp
from scipy.integrate import simps
import pandas as pd
import plotly as px
import plotly.subplots
```

```python
# Входные данные
CONST_N = [4, 7, 11]
CONST_λ = [1/300, 1/400]
```

```python
def fun_p(t):
    return 1 - pow((1 - exp(-l * t)), n)
```

```python
# Расчет показателей
arr = list()
for n in CONST_N:
    for l in CONST_λ:
        # последовательное
        _λ = l * n
        T_c = 1/_λ
        t = np.arange(1,3*T_c)
        p_c = pow(exp(1), (-_λ*t))
        q_c = 1 - p_c
        # параллельное
        i = 0
        while fun_p(i) > 0.04:
            i += 1
        x = np.arange(0, i)
        y = [fun_p(i) for i in x]
        TT_c = simps(y, x)
        tt = np.arange(1, 3 * TT_c)
        pp_c = 1 - pow((1 - pow(exp(1), (-l * tt))), n)
        qq_c = 1 - pp_c
        d1 = {'Время, час': t, 'p(t) послед.': p_c, 'q(t) послед.': q_c}
        d2 = {'Время, час': tt, 'p(t) парал.': pp_c, 'q(t) парал.': qq_c}
        df1 = pd.DataFrame.from_dict(d1, orient='index')
        df2 = pd.DataFrame.from_dict(d2, orient='index')
        arr.append({'Tc':T_c, 'TTc': TT_c, 'n': n, 'λ': l, 'const_data': df1.transpose(), 'paral_data': df2.transpose()})
```

```python
# Построение графиков
fig = px.subplots.make_subplots(x_title= 't', y_title='p(t)')
for i in range(len(arr)):
  fig.add_scatter(y = arr[i]['const_data']['p(t) послед.'], name = f'n = {arr[i]["n"]}, λ = {arr[i]["λ"]:0.4f}')
fig.update_layout(title= 'p(t) const')
fig.show()

fig = px.subplots.make_subplots(x_title= 't', y_title='q(t)')
for i in range(len(arr)):
  fig.add_scatter(y=arr[i]['const_data']['q(t) послед.'], name = f'n = {arr[i]["n"]}, λ = {arr[i]["λ"]:0.4f}')
fig.update_layout(title= 'q(t) const')
fig.show()

fig = px.subplots.make_subplots(x_title= 't', y_title='p(t)')
for i in range(len(arr)):
  fig.add_scatter(y=arr[i]['paral_data']['p(t) парал.'], name = f'n = {arr[i]["n"]}, λ = {arr[i]["λ"]:0.4f}')
fig.update_layout(title= 'p(t) paral')
fig.show()

fig = px.subplots.make_subplots(x_title= 't', y_title='q(t)')
for i in range(len(arr)):
  fig.add_scatter(y=arr[i]['paral_data']['q(t) парал.'], name = f'n = {arr[i]["n"]}, λ = {arr[i]["λ"]:0.4f}')
fig.update_layout(title= 'q(t) paral')
fig.show()
```
